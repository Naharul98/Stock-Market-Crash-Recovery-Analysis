```python
import yfinance as yf
import pandas as pd
```

```python
ftse_100_constituents = pd.read_csv('C:\\Users\\NAHARUL\\Desktop\\FTSE Analysis Naharul\\FTSE 100 Constituents.csv')
ftse_250_constituents = pd.read_csv('C:\\Users\\NAHARUL\\Desktop\\FTSE Analysis Naharul\\FTSE 250 Constituents.csv')
ftse_350_constituents = pd.read_csv('C:\\Users\\NAHARUL\\Desktop\\FTSE Analysis Naharul\\FTSE 350 Constituents.csv')
ftse_aim_100_constituents = pd.read_csv('C:\\Users\\NAHARUL\\Desktop\\FTSE Analysis Naharul\\FTSE AIM 100 Constituents.csv')
```

```python
ftse_100_constituents.set_index('Symbol', inplace=True)
ftse_250_constituents.set_index('Symbol', inplace=True)
ftse_350_constituents.set_index('Symbol', inplace=True)
ftse_aim_100_constituents.set_index('Symbol', inplace=True)
```

```python
for share, x in ftse_100_constituents.iterrows():
    ticker = yf.Ticker(share)
    try:
        for key in ticker.info:
            ftse_100_constituents.at[share, key] = str(ticker.info[key])
    except Exception as e:
        print(share + str(e))
        
for share, x in ftse_250_constituents.iterrows():
    ticker = yf.Ticker(share)
    try:
        for key in ticker.info:
            ftse_250_constituents.at[share, key] = str(ticker.info[key])
    except Exception as e:
        print(share + str(e))
```

```python
for share, x in ftse_350_constituents.iterrows():
    ticker = yf.Ticker(share)
    try:
        for key in ticker.info:
            ftse_350_constituents.at[share, key] = str(ticker.info[key])
    except Exception as e:
        print(share + str(e))
        
for share, x in ftse_aim_100_constituents.iterrows():
    ticker = yf.Ticker(share)
    try:
        for key in ticker.info:
            ftse_aim_100_constituents.at[share, key] = str(ticker.info[key])
    except Exception as e:
        print(share + str(e))
```
